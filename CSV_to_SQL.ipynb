```python
import pandas as pd
import numpy as np
import sqlite3
from sqlalchemy import create_engine
engine = create_engine('sqlite:///hist_trips.sqlite', echo=False)
```

```python
# Store filepath in a variable
file_one = "data/Hist_Divvy_Trips_2018_Q4.csv"
file_two = "data/Hist_Divvy_Trips_2019_Q1.csv"
file_three = "data/Hist_Divvy_Trips_2019_Q2.csv"
file_four = "data/Hist_Divvy_Trips_2019_Q3.csv"
station = "data/Divvy_Bicycle_Stations.csv"
```

```python
file_one_df = pd.read_csv(file_one, encoding="ISO-8859-1")
file_two_df = pd.read_csv(file_two, encoding="ISO-8859-1")
file_three_df = pd.read_csv(file_three, encoding="ISO-8859-1")
file_four_df = pd.read_csv(file_four, encoding="ISO-8859-1")
stations_df = pd.read_csv(station, encoding="ISO-8859-1")
```

```python
vertical_stack_df = pd.concat([file_one_df, file_two_df], axis=0)
```

```python
vertcl_stck_df = pd.concat([vertical_stack_df, file_three_df], axis=0)
```

```python
hist_divvy_df = pd.concat([vertcl_stck_df, file_four_df], axis=0)
```

```python
hist_divvy_df.count()
```

```python
hist_divvy_df
```

```python
hist_divy_df = hist_divvy_df.dropna()
```

```python
len(hist_divy_df)
```

```python
hist_dvy_df = hist_divy_df.drop(["bikeid", "from_station_name", "to_station_name"], axis=1)
```

```python
len(hist_dvy_df)
```

```python
hist_dvy_df
```

```python
hist_dvy_df.index = range(len(hist_dvy_df.index))
```

```python
hst_dvy_df = hist_dvy_df[hist_dvy_df.usertype == "Subscriber"]
```

```python
hst_dvy_df.head()
```

```python
hst_dvy_df.to_sql('hist_trips', con=engine)
```

```python
stations_df.to_sql('stations', con=engine)
```
