```python
# Import Dependencies
import pandas as pd
import numpy as np
import sqlite3
from sqlalchemy import create_engine
engine = create_engine('sqlite:///data/hist_trips.sqlite', echo=False)
```

```python
# Store filepath in a variable
file_one = "data/Hist_Divvy_Trips_2019_Q2.csv"
station = "data/Divvy_Bicycle_Stations.csv"
```

```python
# Create Data frames from csv files
hist_trips_divvy_df = pd.read_csv(file_one, encoding="ISO-8859-1")
stations_df = pd.read_csv(station, encoding="ISO-8859-1")
```

```python
# Drop rows with NAN's
hist_trps_divvy_df = hist_trips_divvy_df.dropna()
```

```python
# Remove columns you don't need
hist_trps_divy_df = hist_trps_divvy_df.drop(["bikeid", "from_station_name", "to_station_name"], axis=1)
```

```python
hist_trps_dvy_df = hist_trps_divy_df[hist_trps_divy_df.usertype == "Subscriber"]
```

```python
hist_trps_dvy_df.to_sql('hist_trips', con=engine)
```
